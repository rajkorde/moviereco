```python
import azureml.core

print("SDK version:", azureml.core.VERSION)
```

```python
from azureml.core import Workspace

ws = Workspace.from_config()
print(ws.name, ws.resource_group, ws.location, ws.subscription_id, sep = '\n')
```

```python
from azureml.core import Experiment
experiment_name = 'train-on-amlcompute'
experiment = Experiment(workspace = ws, name = experiment_name)
```

```python
project_folder = '.'
```

```python
ws.get_default_compute_target(type='CPU')
```

```python
ws.compute_targets
```

```python
from azureml.core.runconfig import RunConfiguration
from azureml.core.conda_dependencies import CondaDependencies
from azureml.core.runconfig import DEFAULT_CPU_IMAGE
```

```python
run_config = RunConfiguration(framework='python')

run_config.target = ws.compute_targets['prodmod-compute1'].name
```

```python
run_config.environment.docker.enabled = True
#run_config.environment.docker.base_image = DEFAULT_CPU_IMAGE
```

```python
run_config.environment.python.user_managed_dependencies = False
run_config.environment.python.conda_dependencies = CondaDependencies.create(conda_packages=['scikit-learn'])
```

```python
from azureml.core.script_run_config import ScriptRunConfig
```

```python
script_run_config = ScriptRunConfig(source_directory=project_folder, script='train.py', run_config=run_config)
```

```python
run = experiment.submit(config=script_run_config)
```

```python
%%time

run.wait_for_completion(show_output=True)
```
